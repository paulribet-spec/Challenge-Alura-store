### Importación de datos



```python
import matplotlib.pyplot as plt
```

```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
ingresos_Tienda1 = tienda["Precio"].sum()
ingresos_Tienda2 = tienda2["Precio"].sum()
ingresos_Tienda3 = tienda3["Precio"].sum()
ingresos_Tienda4 = tienda4["Precio"].sum()
```

```python
conteo_tienda1 = tienda["Categoría del Producto"].value_counts()
conteo_tienda2 = tienda2["Categoría del Producto"].value_counts()
conteo_tienda3 = tienda3["Categoría del Producto"].value_counts()
conteo_tienda4 = tienda4["Categoría del Producto"].value_counts()
```

# 3. Calificación promedio de la tienda


```python
Calificacion_Tienda1 = tienda["Calificación"].mean()
Calificacion_Tienda2 = tienda2["Calificación"].mean()
Calificacion_Tienda3 = tienda3["Calificación"].mean()
Calificacion_Tienda4 = tienda4["Calificación"].mean()
```

```python
print(Calificacion_Tienda1)
print(Calificacion_Tienda2)
print(Calificacion_Tienda3)
print(Calificacion_Tienda4)
```

# 4. Productos más y menos vendidos

```python
Productos_tienda1 = tienda["Producto"].value_counts()
Productos_tienda2 = tienda2["Producto"].value_counts()
Productos_tienda3 = tienda3["Producto"].value_counts()
Productos_tienda4 = tienda4["Producto"].value_counts()
```

# 5. Envío promedio por tienda

**¿Qué tienda logra vender más y mantener clientes satisfechos con menor fricción para comprar?**

##Tabla resumen
Generaremos una tabla resumen de los datos para facilitar la visualización de estos

```python
comparacion = pd.DataFrame({
    "Ventas Totales": [
        ingresos_Tienda1,
        ingresos_Tienda2,
        ingresos_Tienda3,
        ingresos_Tienda4
    ],
    "Satisfacción Promedio": [
        Calificacion_Tienda1,
        Calificacion_Tienda2,
        Calificacion_Tienda3,
        Calificacion_Tienda4
    ],
    "Costo Envío Promedio": [
        CostoEnvio_promedioT1,
        CostoEnvio_promedioT2,
        CostoEnvio_promedioT3,
        CostoEnvio_promedioT4
    ]
},
index=["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"])
comparacion
```

```python
#normalizamos los datos:
comparacion_norm = (comparacion - comparacion.min()) / (comparacion.max() - comparacion.min())
comparacion_norm
```

```python
comparacion_norm.plot(kind="bar", figsize=(10,5))
plt.title("Comparación normalizada entre tiendas")
plt.show()
```

Podemos notar que no hay una tendencia entre la cantidad de ingreso por venta total de cada tienda y la satisfacción promedio del cliente por tienda, por lo tanto la experiencia del cliente no es el principal impulsor comercial.

##Productos más vendidos por tienda

Quiero saber cuales son las categorias más vendidas independiente de la tienda:

De esta grafica podemos observar:

- En 3 de las 4 categorias más vendidas la tienda 2 es la que menos ingresos genera por venta.
- La tienda 3 y 4 lideran en 1 categoria en solitario y en 1 empatan.

Analizaremos la estabilidad del modelo de ingresos.

Calculo de precio promedio por tienda y desviación estándar del precio:

```python
precio_T1 = tienda["Precio"].mean()
precio_T2 = tienda2["Precio"].mean()
precio_T3 = tienda3["Precio"].mean()
precio_T4 = tienda4["Precio"].mean()
```

```python
std_T1 = tienda["Precio"].std()
std_T2 = tienda2["Precio"].std()
std_T3 = tienda3["Precio"].std()
std_T4 = tienda4["Precio"].std()
```

Tabla comparativa:

```python
analisis_final = pd.DataFrame({
    "Precio Promedio": [precio_T1, precio_T2, precio_T3, precio_T4],
    "Desviacion Precio": [std_T1, std_T2, std_T3, std_T4]
},
index=["Tienda 1","Tienda 2","Tienda 3","Tienda 4"])
```

```python
analisis_final["Coef_Variacion"] = (
    analisis_final["Desviacion Precio"] /
    analisis_final["Precio Promedio"]
)

analisis_final
```

#Informe Final
El presente análisis tuvo como objetivo evaluar el desempeño comercial de cuatro tiendas, con el fin de determinar si existen diferencias significativas en su rendimiento y si estas permiten sustentar una decisión estratégica informada.

Para ello, se analizaron variables clave como ingreso total por tienda, ingreso promedio por venta, liderazgo por categorías, satisfacción promedio de clientes y estabilidad del modelo de precios.
A partir del análisis realizado se concluye que **no existe relación entre el nivel de satisfacción promedio y el ingreso total por tienda**, por lo que la satisfacción no explica las diferencias en desempeño comercial.

Dado que **la cantidad total de ventas es igual en todas las tiendas**, las diferencias de ingreso se explican exclusivamente por el ingreso promedio por venta.

En este contexto:

La Tienda 1 presenta el mayor ingreso promedio por venta.

La Tienda 2 muestra bajo desempeño relativo en 3 de las 4 categorías más vendidas.

Las Tiendas 3 y 4 lideran categorías específicas, pero no de forma dominante.

El análisis de variabilidad (coeficiente de variación) permite evaluar la estabilidad del modelo de ingresos.

Por lo tanto, la decisión estratégica debe considerar:

Nivel de ingreso promedio por venta, liderazgo por categorías, estabilidad del modelo de precios y posicionamiento competitivo relativo

Si el objetivo es maximizar rentabilidad con estabilidad, la tienda con **menor ingreso promedio y mayor variabilidad** sería la candidata menos eficiente resultando en la **Tienda 4.**

```python
plt.scatter(
    analisis_final["Precio Promedio"],
    analisis_final["Coef_Variacion"]
)

for tienda_nombre in analisis_final.index:
    plt.text(
        analisis_final.loc[tienda_nombre,"Precio Promedio"],
        analisis_final.loc[tienda_nombre,"Coef_Variacion"],
        tienda_nombre
    )

plt.xlabel("Precio Promedio")
plt.ylabel("Coeficiente de Variación")
plt.title("Estabilidad del modelo comercial")

plt.show()
```
